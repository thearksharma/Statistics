```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
print("Comparision of Mean & Median\n")
import numpy as np
arr = np.array([20,30,25,45,300])
# with outliers in dataset
print(f"Mean: {arr.mean()} & Median: {np.median(arr)}")
arr = np.array([-50,20,30,25,45])
print(f"Mean: {arr.mean()} & Median: {np.median(arr)}")
```

##Left - skewed

```python
y = np.array([40,40,40,38,38,35,35,35,35,31,31,31,31,31,29,29,29,29,29,29,29,29,27,27,21,21,17,17,7,0])

fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 8))

sns.kdeplot(y, ax=axes[0, 0])
axes[0, 0].set_title('Original y')

sns.kdeplot(y**2, ax=axes[0, 1])
axes[0, 1].set_title('y^2')

sns.kdeplot(y**3, ax=axes[1, 0])
axes[1, 0].set_title('y^3')

sns.kdeplot(np.exp(y), ax=axes[1, 1])
axes[1, 1].set_title('exp(y)')

plt.tight_layout()
plt.show()
```

##Right - Skewed

```python
x = np.array([1,1,1,2,2,4,4,4,4,6,6,6,8,8,8,8,8,9,9,9,9,9,13,17,20,20,25,25,33,39])

fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(15, 8))

sns.kdeplot(x, ax=axes[0, 0])
axes[0, 0].set_title('Original x')

sns.kdeplot(np.log(x), ax=axes[0, 1])
axes[0, 1].set_title('log(x)')

sns.kdeplot(x*-1, ax=axes[0, 2])
axes[0, 2].set_title('-x')

sns.kdeplot((x)**0.5, ax=axes[1, 0])
axes[1, 0].set_title('x^0.5')

sns.kdeplot((x)**(1/3), ax=axes[1, 1])
axes[1, 1].set_title('x^(1/3)')

sns.kdeplot(1/x, ax=axes[1, 2])
axes[1, 2].set_title('1/x')

plt.tight_layout()
plt.show()
```

```python
data = np.random.normal(loc=10, scale=10, size=10)

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))

sns.kdeplot(data, ax=axes[0])
axes[0].set_title('KDE Plot')

axes[1].hist(data)
axes[1].set_title('Histogram')

plt.tight_layout()
plt.show()
```
